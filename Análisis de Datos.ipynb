## Media
Como primer operación a realizar, se obtendrá el promedio de la columna anteriormente mencionada

```python
print('Promedio/Media: ', df['score'].mean())
```

La calificación promedio que hay entre estas universidades es de *47.8* aproximadamente.
Para saber más acerca de países que están entre el 40 y 50 de calificación, se hace uso de un **filtro**:

```python
enPromedio = df[(df['score'] >= 40) & (df['score'] <= 50)]
enPromedio
```

```python
len(enPromedio)
```

En conjunto con la función *len*, se puede conocer que ***1868*** universidades de las ***2200*** están en el rango intermedio.
